# Gradient Boosting Regression 

```python
import pandas as pd
df=pd.DataFrame({
    'iq':[90,100,110,120,80],
    'cgpa':[8,7,6,9,5],
    'salary':[3,4,8,6,3]
})
df.head()
```

```python

import numpy as np
base_model=np.mean(df['salary'])
```

```python
df['residual1']=df['salary']-base_model
```

```python
df['residual1']
```

```python
from sklearn.tree import DecisionTreeRegressor
model=DecisionTreeRegressor()
```

```python
df['base_model']=base_model
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df[['iq','cgpa']],df.residual1,test_size=0.2,random_state=42)
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_train,y_train)
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
#fig,ax=plt.subplot(fisize=(10,5))
plt.figure(figsize=(10,5))
tree.plot_tree(model, filled=True)
plt.show()
```
