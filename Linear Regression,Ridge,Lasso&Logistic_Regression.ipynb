```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.datasets import fetch_california_housing
```

```python
df=fetch_california_housing()          # Importing dataset
df
```

```python
dataset=pd.DataFrame(df.data)   # creating a dataframe
```

```python
dataset.columns=df.feature_names             #assigning column names from feature_names from the dataset- Dependent features
dataset.head()
```

```python
dataset['Price']=df.target                    # Independent feature addition to the newly created dataset
```

```python
dataset.head()
```

# If i want to solve linear regression, divide the dataset into dependent and independent features

```python
X=dataset.iloc[:,:-1]      # Just skip the last feature with this code- first : referes records second one columns Independent feature
y=dataset.iloc[:,-1]       #Dependent feature
```

```python
X
```

```python
y
```

```python
# Linear Regression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
lin_reg=LinearRegression()
mse=cross_val_score(lin_reg,X,y,scoring='neg_mean_squared_error',cv=5)
mean_mse=np.mean(mse)
print(mean_mse)
```

```python
#Ridge Regression
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
ridge=Ridge()
params={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20]}
ridge_regressor=GridSearchCV(ridge,params,scoring='neg_mean_squared_error',cv=5)
ridge_regressor.fit(X,y)
```

```python
print(ridge_regressor.best_params_)  #Selecting best parameters from the alpha
print(ridge_regressor.best_score_)   # Best Score
```

```python
#Lasso Regression
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
lasso=Lasso()
params={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,45,50,100]}
lasso_regressor=GridSearchCV(lasso,params,scoring='neg_mean_squared_error',cv=5)
lasso_regressor.fit(X,y)
```

```python
print(lasso_regressor.best_params_)  #Selecting best parameters from the alpha
print(lasso_regressor.best_score_)   # Best Score
```

Take up train and test data and execute the above algorithms again

```python
# Linear Regression on train and test datasets
from sklearn.model_selection import train_test_split
X_test,X_train,y_test,y_train=train_test_split(X,y,test_size=.33,random_state=42)
```

```python
#Ridge Regression
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
ridge=Ridge()
params={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20]}
ridge_regressor=GridSearchCV(ridge,params,scoring='neg_mean_squared_error',cv=5)
ridge_regressor.fit(X_test,y_test)
```

```python
print(ridge_regressor.best_params_)  #Selecting best parameters from the alpha
print(ridge_regressor.best_score_)   # Best Score
```

```python
# Lasso Regression on train and test datasets
from sklearn.model_selection import train_test_split
X_test,X_train,y_test,y_train=train_test_split(X,y,test_size=.33,random_state=42)
```

```python
#Lasso Regression
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
lasso=Lasso()
params={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,45,50,100]}
lasso_regressor=GridSearchCV(lasso,params,scoring='neg_mean_squared_error',cv=5)
lasso_regressor.fit(X_test,y_test)
```

```python
print(lasso_regressor.best_params_)  #Selecting best parameters from the alpha
print(lasso_regressor.best_score_)   # Best Score
```
